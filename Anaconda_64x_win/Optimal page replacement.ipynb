```python
def optimal_page_replacement(pages, capacity):
    memory = []
    page_faults = 0

    print("\nOptimal Page Replacement Simulation\n")

    for i in range(len(pages)):
        page = pages[i]

        if page not in memory:
            page_faults += 1

            if len(memory) < capacity:
                memory.append(page)
            else:
                # Find the page that wonâ€™t be used for the longest time in future
                future_uses = []
                for m in memory:
                    if m not in pages[i+1:]:
                        future_uses.append(float('inf'))
                    else:
                        future_uses.append(pages[i+1:].index(m))

                index_to_replace = future_uses.index(max(future_uses))
                removed = memory[index_to_replace]
                memory[index_to_replace] = page
                print(f"Removed Page: {removed}")

            print(f"Inserted Page: {page} -> Memory: {memory}")
        else:
            print(f"Page {page} already in memory -> No Page Fault")

    print("\nTotal Page Faults:", page_faults)


# Example Input
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
capacity = 3

optimal_page_replacement(pages, capacity)
```
