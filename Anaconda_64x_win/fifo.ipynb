```python
# FIFO Page Replacement Algorithm Simulation

def fifo_page_replacement(pages, capacity):
    memory = []      # List to store current pages in memory
    page_faults = 0  # Counter for page faults

    print("\nFIFO Page Replacement Simulation\n")

    for page in pages:
        # If page is not in memory -> page fault occurs
        if page not in memory:
            page_faults += 1

            # If memory is full, remove the oldest page (first inserted)
            if len(memory) == capacity:
                removed = memory.pop(0)
                print(f"Removed Page: {removed}")

            memory.append(page)
            print(f"Inserted Page: {page} -> Memory: {memory}")
        else:
            print(f"Page {page} already in memory -> No Page Fault")

    print("\nTotal Page Faults:", page_faults)


# Example Input
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
capacity = 3

fifo_page_replacement(pages, capacity)
```
