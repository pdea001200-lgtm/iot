```python
# Program to simulate First Fit memory allocation

def first_fit(block_sizes, process_sizes):
    n = len(block_sizes)       # Number of memory blocks
    m = len(process_sizes)     # Number of processes

    # Store the block number allocated to each process (-1 means not allocated)
    allocation = [-1] * m

    # Traverse each process
    for i in range(m):
        # Traverse each block
        for j in range(n):
            # If block j can accommodate process i
            if block_sizes[j] >= process_sizes[i]:
                # Allocate block j to process i
                allocation[i] = j
                # Reduce the available size of block j
                block_sizes[j] -= process_sizes[i]
                # Once allocated, break out of the inner loop
                break

    # Display the allocation result
    print("\nProcess No.\tProcess Size\tBlock No.")
    for i in range(m):
        print(f"{i + 1}\t\t{process_sizes[i]}\t\t", end='')
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")


# ---------- MAIN PROGRAM ----------
if __name__ == "__main__":
    # Input memory block sizes
    block_sizes = [100, 500, 200, 300, 600]
    # Input process sizes
    process_sizes = [212, 417, 112, 426]

    print("Memory Blocks:", block_sizes)
    print("Process Sizes:", process_sizes)

    first_fit(block_sizes.copy(), process_sizes)
```
