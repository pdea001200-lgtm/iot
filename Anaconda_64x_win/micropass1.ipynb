```python
class MNTEntry:
    def __init__(self, i, n, mdt_i):
        self.index, self.name, self.mdt_index = i, n, mdt_i

MNT, MDT, ALA = [], [], {}

def pass1(prog):
    in_macro, name, mnt_i, formals = False, "", 0, []
    for line in prog:
        line = line.strip()
        if not line: continue
        p = line.replace(",", " ").split()

        if line == "MACRO": in_macro = True; formals.clear(); continue
        if in_macro:
            if line == "MEND": MDT.append("MEND"); in_macro = False; continue
            if not name:
                name = p[0]; MNT.append(MNTEntry(mnt_i, name, len(MDT))); mnt_i += 1; formals = p[1:]
            else:
                MDT.append(" ".join(f"#{formals.index(w)}" if w in formals else w for w in p))
        else:
            for e in MNT:
                if e.name == p[0]: ALA[p[0]] = p[1:]

def show():
    print("MNT:\nIdx\tName\tMDT")
    [print(f"{m.index}\t{m.name}\t{m.mdt_index}") for m in MNT]
    print("\nMDT:")
    [print(f"{i}\t{d}") for i, d in enumerate(MDT)]
    print("\nALA:")
    [print(f"{k}: {v}") for k, v in ALA.items()]

prog = [
    "MACRO","INCR &A,&B","ADD AREG,&A","SUB BREG,&B","MEND",
    "MACRO","DECR &X,&Y","SUB AREG,&X","SUB BREG,&Y","MEND",
    "START 100","INCR A,B","DECR 5,10","END"
]

pass1(prog); show()
```
